```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_roots = 17 # Number of roots based on ionization threshold
savepath = "/home/sanchez/Trieste_Project/Chemistry/Scripts/article_xas_xncd/data/"
```

## Generate PMM XCD spectrum Carbon K-Edge, Perfluorohexane, CAM-B3LYP

```python
main_path1 = "/home/sanchez/Trieste_Project/Chemistry/Scripts/article_xas_xncd/data/spettri-singoli/PMM-CARBON-K-C6F14/"

spectrum1 = 0
energies1 = 0

for i in range(1, n_roots+1):
    data_spectrum = np.genfromtxt(main_path1 + f"ecd-0-{i}-core")
    spectrum1 += data_spectrum[:,1]
    energies1 = data_spectrum[:,0]

np.savetxt(savepath + "spettro-cam-perfluoro", np.column_stack((energies1, spectrum1)))

plt.plot(energies1, spectrum1)
plt.xlabel("Energy (eV)")
plt.ylabel("XCD (a.u.)")
plt.xlim(275, 282)
plt.show()
```
